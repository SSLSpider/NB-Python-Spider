```python
def readURL(URL,heads):
    req = urllib2.Request(URL,headers=heads)
    resp = urllib2.urlopen(req)
    respData = resp.read()
    soup = BeautifulSoup(respData, "html.parser")
    return soup
```

```python
def Get_Title(soup):
    hs=soup.find_all(['h1','h2','h3','h4'])
    h_list=[]
    for h in hs :
        label = h.text
        LL=[h for key in Education_keys if key in label.lower()]
        if len(LL) is not 0:
            h_list.append(LL[0])
    return h_list
```

```python
def Education_Exaction(H_list):
    Education_list=[]
    for h in H_list:
        a=h.find_next_siblings()
        Part_tmp=[ b.text for b in a]
        Education_list.append(Part_tmp[0])
    return Education_list
```
