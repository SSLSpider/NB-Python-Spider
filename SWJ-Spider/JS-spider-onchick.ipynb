```python
def readString(target_name):
    tittle=[]
    for text in target_name:
        if text.string is not None:
            tittle.append(text.string)
    return tittle
```

```python
No=[]
target=[]
department=[]
datetime=[]
explain_text=[]
tittle_name=[]
tittle_tmp=[]
for url in tqdm.tqdm(url_list):
    tittle=[]
    rep=urllib2.Request(url)
    reqs=urllib2.urlopen(rep)
    readData=reqs.read()
    soup=BeautifulSoup(readData,'html.parser')
    
    target_name=soup.find_all('td',attrs={"align":"center"})
    td_xx=soup.find_all('td',attrs={"class":"xzcf_xx"})
   
    tittle=readString(target_name)
    tittle_name.append(tittle[0])
    tittle_tmp.append(tittle)
    
    No.append(td_xx[0].text)
    target.append(td_xx[1].text)
    department.append(td_xx[2].text)
    datetime.append(td_xx[3].text)
    explain_text.append(soup.find_all('td',attrs={'class':'xzcf_jds'})[0].text)
```
