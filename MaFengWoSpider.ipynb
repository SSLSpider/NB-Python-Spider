```python
from bs4 import BeautifulSoup as Soup
from urllib.request import urlopen
import re


def getlinks(Url):
    html = urlopen(Url)
    bsObj = Soup(html, 'html5lib')
    post = bsObj.find('div', {'class':'gl-post'})
    return post.findAll('a', {'href':re.compile("^(http|https)://www\.mafengwo\.cn/i/.*$")})

def printContent(url):
    html = urlopen(url)
    bsObj = Soup(html, 'html5lib')
    contents = bsObj.find('div', {'class':'main'}).findAll({'h1', 'h2', 'p'})
    for cont in contents:
        print(cont.get_text())
    print('\n**-----------------------------------------------------------------**\n')
    pass


links = getlinks("http://www.mafengwo.cn/gonglve/")
print(len(links))
for link in links:
    if 'href' in link.attrs:
        printContent(link.attrs['href'])
```
